#### Metoda reguła Falsi

```julia
plot(f1, -1, 1)
```

```julia
function falsi_wrap(f, a, b)
    root = find_zero(f, (a, b), FalsePosition(), verbose=true) 
    if f(root) == 0
        print("Success.\n")
    else
        print("Failure.\n")
        if (f(prevfloat(root)) > 0 && f(nextfloat(root)) < 0)
            print(" But as close as could be!\n")
        end
    end
    print(root)
end
```

#### Metoda Newtona

```julia
# definujemy D(f) obliczającą  funkcje pochodną 
D(f) = x->ForwardDiff.derivative(f, float(x))
```

```julia
function wrap_newton(f)
    root = find_zero((f, D(f)),0, Roots.Newton(),verbose=true)
    if f(root) == 0
        print("Success.\n")
    else
        print("Failure.\n")
        if (f(prevfloat(root)) > 0 && f(nextfloat(root)) < 0)
            print(" But as close as could be!\n")
        end
    end
    print(root)
end
```

```julia
wrap_newton(f1)
```

```julia
wrap_newton(f2)
```

```julia
wrap_newton(f3)
```

```julia
wrap_newton(f4)
```

```julia
wrap_newton(f5)
```

```julia
wrap_newton(f6)
```

#### Metoda Steffensena

```julia
function wrap_steffensen(f)
    root = x = find_zero(f, 0, Order2(), verbose=true) 
    if f(root) == 0
        print("Success.\n")
    else
        print("Failure.\n")
        if (f(prevfloat(root)) > 0 && f(nextfloat(root)) < 0)
            print(" But as close as could be!\n")
        end
    end
    print(root)
end
```

```julia
wrap_steffensen(f1)
```

```julia
wrap_steffensen(f3)
```

```julia
wrap_steffensen(f4)
```

```julia
wrap_steffensen(f5)
```

```julia
wrap_steffensen(f6)
```

##### Metoda Newtona
Metoda Newtona nie działa, kiedy pierwszy punkt który podajemy jest w jakimś ekstremum, wtedy nachylenie jest zerowe i przybliżenie do korzenia jest niemożliwe.

```julia
fun2(x) = x^3 -3x + 3
```

```julia
plot(fun2, -3, 3)
```

```julia
root = find_zero((fun2, D(fun2)),1, Roots.Newton(),verbose=true)
if f(root) == 0
    print("Success.\n")
else
    print("Failure.\n")
    if (fun2(prevfloat(root)) > 0 && fun2(nextfloat(root)) < 0)
        print(" But as close as could be!\n")
    end
end
print(root)
```

##### Metoda Steffensena
Słabą stroną metody Steffensena jak i metody Newtona jest wybranie punktu początkowego, jeżeli on jest zadaleko, to kolejne x-y nie rozbiegają się w niskończoność.

```julia
fun3(x) = tan(x)^3
```

```julia
plot(fun3, -1, 1)
```

```julia
root = x = find_zero(fun3, 2, Order2(), verbose=true) 
if fun3(root) == 0
    print("Success.\n")
else
    print("Failure.\n")
    if (fun3(prevfloat(root)) > 0 && fun3(nextfloat(root)) < 0)
        print(" But as close as could be!\n")
    end
end
print(root)
```
