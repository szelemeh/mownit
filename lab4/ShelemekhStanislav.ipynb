```r
# wczytywanie danych

df = read.csv("results.csv")
df
```

```r
# wyznaczenie średnich czasów dla takich samych funkcji i rozmiarów

avg_time = aggregate( time ~ size:only_vectors, data=df, FUN=mean)
avg_time
```

```r
# dołaczenie kolumny z odchyleniem standardowym

avg_time$sd = aggregate( time ~ size:only_vectors, data=df, FUN=sd)$time
avg_time
```

```r
vector_vector = avg_time[avg_time$only_vectors==1,]
matrix_vector = avg_time[avg_time$only_vectors==0,]
vector_vector
matrix_vector
```

```r
# sporządzenie wykresów srednich czasów i ich odchyleń standardowych

library("ggplot2")

vector_errorbar = geom_errorbar( 
    data=vector_vector, mapping=aes(ymin=time-sd, ymax=time+sd), width=1, color="blue")
matrix_errorbar = geom_errorbar( 
    data=matrix_vector, mapping=aes(ymin=time-sd, ymax=time+sd),width=1, color="blue")
            
ggplot(vector_vector, aes(size, time)) + geom_point() + vector_errorbar                                
ggplot(matrix_vector, aes(size, time)) + geom_point() + matrix_errorbar
                                
```

```r
# aproksymacja wielomianowa

x = vector_vector$size
y = vector_vector$time
projection1 = data.frame(x, y)
projection1
```

```r
# dopasowanie wielomianem stopnia 1

fit = lm(y ~ poly(x, 1, raw=TRUE), data=projection1)
new_projection1 = data.frame(x = seq(0,1000000, length.out=250))

# generowanie krzywej wielomianu

new_projection1$y = predict(fit, new_projection1)
```

```r
# wykres punktów

ggplot(projection1,aes(x,y)) + geom_point()
```

```r
# dodanie do wykresu krzywej wielomianu

last_plot() + geom_line(data=new_projection1, aes(x,y))
```
